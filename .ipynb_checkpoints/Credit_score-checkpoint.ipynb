# Problem Statement

Obtaining loans from financial institutions like banks has become a common phenomenon in the past few years and the risks associated with making a decision on loan approval is immense.Every year there are cases of individuals not repaying their loan amounts which results in huge financial losses for banks. The aim of the project is to identify credit worthy customers while minimizing their risk of defaulting on their loans.

The project aims at predicting whether a person will default on a loan or not given some variables.This is a binary classification problem.

```python
# extract day, month and year from StartDate to find out which month and dates customers applied for loans the most

def extract_date(loans_df, column):
    loans_df[column+"_day"] = loans_df[column].apply(lambda x: x.day)
    loans_df[column+"_month"] = loans_df[column].apply(lambda x: x.month)
    loans_df[column+"_year"] = loans_df[column].apply(lambda x: x.year)
    
extract_date(loans_df, 'DisbursementDate')
```

# Bivariate Analysis

```python
#Binning the Amounts in order to find out the paying patterns for different amount groups
bins = [0, 10000, 50000, 100000, 200000]
loans_df['AmountGroups'] = pd.cut(loans_df['Amount'], bins,include_lowest = True,labels = ['0-10000','10001-50000','50001-100000','100000-200000'])
loans_df.head(20)
```

```python
print(pd.crosstab(loans_df['AmountGroups'],loans_df['Default']))
LoanAmount_bin=pd.crosstab(loans_df['AmountGroups'],loans_df['Default']) 
LoanAmount_bin.div(LoanAmount_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.title('Proportion of non-defaulters to defaultes in the amount groups')
plt.xlabel('LoanAmount') 
P = plt.ylabel('Percentage')
```

The proportion of the different amount groups is more or less same for both the defaulted and paid loans.This contradicts the hypothesis that if the loan amount is lower, the chances of paying loans is higher. From this data it can be inferred that amout does not necessarily affect if the customer will default or not.

```python
print(loans_df.groupby('Default')['Amount'].mean())

loans_df.groupby('Default')['Amount'].mean().plot.bar()
plt.title('Mean Loan amount between defaulters and non-defaulters')
plt.ylabel('Mean Amount')
plt.xlabel('Default')
```

We dont see any significant difference in the mean loan Amount between the defaulters and non-defaulters. (9290 vs 9150).

```python
print(loans_df.groupby('Default')['Age'].mean())

loans_df.groupby('Default')['Age'].mean().plot.bar()
plt.title('Mean Age between defaulters and non-defaulters')
plt.ylabel('Mean Age')
plt.xlabel('Default')
```

The mean age of defaulters and non-defaulters is almost similar, that is, 34 for Defaulters and 38 for non-defaulters. Binning the ages in different groups might give us more insight.

```python
bins = [0, 25, 50, 75, 100, 120]
loans_df['AgeGroups'] = pd.cut(loans_df['Age'], bins,include_lowest = True,labels = ['Under25','Under50','Under75','Under100','Under120'])
loans_df.head(20)
```

```python
print(pd.crosstab(loans_df['AgeGroups'],loans_df['Default']))
Age_bin=pd.crosstab(loans_df['AgeGroups'],loans_df['Default']) 
Age_bin.div(Age_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.title('Proportion of non-defaulters to defaultes in the age groups')
plt.xlabel('Age') 
P = plt.ylabel('Percentage')
```

The proportion of non-defaulters to defaulters is more or less similar for the different age groups. Proportion of defaulters is however slightly higher for customers under 25. This Might be attributed to alot of them being students and hence unemployed.

# FEATURE ENGINEERING

```python
#To find the number of times each customer applied for a loan in the past.
loans_df['NoOfLoans'] = loans_df.groupby('Customer_WID')['Customer_WID'].transform('count')
```

```python
loans_df = loans_df.drop(['Count'], axis=1)
loans_df.head(10)
```

```python
#How big the loan a person has taken with respect to his FinancialMeasure1,FinancialMeasures,FinancialMeasure3,FinancialMeasures4,CRBScore to Amount ratio
#Did not use variable age since it doesnt add much weight in determining amount offered to client 
loans_df['loan_to_income'] = loans_df['FinancialMeasure1'] +loans_df['FinancialMeasure2']+loans_df['FinancialMeasure3']+loans_df['FinancialMeasure4']+loans_df['CRBScore']+loans_df['SOR']+loans_df['MonthsSinceActive']+loans_df['MonthsSinceOpen'] /loans_df['Amount']
```

```python
loans_df.head(10)
```

```python
#There is an annual interest rate of 13%. Create a new feature called Interest
loans_df['Interest'] = 13/100 * 1/12 *loans_df['Amount']
loans_df.head(10)
```

```python
#The loan application fee is 5% of the loan amount. Create a new feature called Loan_fees
loans_df['Loan_fees'] = 5/100 * loans_df['Amount']
loans_df.head(10)
```

```python
#Correlation Matrix
corr = loans_df.corr()
corr = (corr)
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.title('Heatmap of Correlation Matrix')
corr
```

The heatmap with the new features added gives some insight. The Higher the number of loans a customer has, the less likely they are to Default on a loan. 

The higher the loan_to_income value a customer has, the higher the loan_amount available.
Loan_to_income also has a very high positive correlation with financial measures 3 and 4.
