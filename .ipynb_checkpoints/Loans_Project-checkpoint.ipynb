```python
#checking dataframe to see which rows have missing values(11 rows)
null_data = loans_df[loans_df.isnull().any(axis=1)]
null_data
```

```python
#percentage of missing value counts in each of these columns
miss = loans_df.isnull().sum()/len(loans_df) * 100
miss = miss[miss > 0]
miss.sort_values(inplace=True)
miss
```

```python
#visualising missing values
miss = miss.to_frame()
miss.columns = ['count']
miss.index.names = ['Name']
miss['Name'] = miss.index

#plot the missing value count
sns.set(style="whitegrid", color_codes=True)
sns.barplot(x = 'Name', y = 'count', data=miss)
plt.xticks(rotation = 90)
sns.plt.show()
```

```python
#checking for outliers to decide how to handle missing values
#Most columns with missing values have outliers hence populating with mean in those columns isnt a good idea.
sns.boxplot(x=loans_df['Age'])
```

```python
#Populate missing values in age column with median. 
median_value=loans_df['Age'].median()
loans_df['Age']=loans_df['Age'].fillna(median_value)
```

```python
mode_value=loans_df['CustomerType'].mode()
loans_df['CustomerType']=loans_df['CustomerType'].fillna(mode_value)
```

```python
loans_df.isnull().any()
```

```python
# Get number of people that defaulted and those that paid fully
pos = loans_df[loans_df["Default"] == 1].shape[0]
neg = loans_df[loans_df["Default"] == 0].shape[0]
print(f"Positive examples = {pos}")
print(f"Negative examples = {neg}")
print(f"Proportion of positive(defaulted) to negative(not defaulted) examples = {(pos / neg) * 100:.2f}%")
plt.figure(figsize=(8, 6))
sns.countplot(loans_df["Default"])
plt.xticks((0, 1), ["Paid fully", "Not paid fully"])
plt.xlabel("")
plt.ylabel("Count")
plt.title("Class counts", y=1, fontdict={"fontsize": 20});
```

The Dataset is imbalanced

```python
bins = [0, 25, 50, 75, 100, 120]
loans_df['AgeGroups'] = pd.cut(loans_df['Age'], bins,include_lowest = True,labels = ['Under25','Under50','Under75','Under100','Under120'])
loans_df.head(20)
```

```python
loans_df.groupby(['AgeGroups', 'Default'] )['Default'].count()
```

```python
loans_df.groupby(['AgeGroups', 'Default'] )['Default'].count().plot.bar(figsize=(8, 6));
```

```python
bins = [0, 10000, 50000, 100000, 200000]
loans_df['AmountGroups'] = pd.cut(loans_df['Amount'], bins,include_lowest = True,labels = ['Under10K','Under50K','Under100K','Under200K'])
loans_df.head(20)
```

```python
loans_df.groupby(['AmountGroups', 'Default'] )['Default'].count()
```
